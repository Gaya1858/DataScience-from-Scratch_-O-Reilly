```python
from matplotlib import pyplot as plt
import numpy as np
import random
```

```python
random.seed(1)
year = sorted([random.randrange(1950, 2020) for _ in range(20)])
```

```python
gdp = [random.randrange(300, 14958) for _ in range(20)] 
gdp1  =['%.1f' % random.uniform(0,1) for _ in range(20)]
```

```python
gdp2 = sorted([a+(float(b)) for a,b in zip(gdp, gdp1)])
```

```python
# create a line chart Years on X axis and gdp2 on y axis
plt.plot(year, gdp2, color="orange",marker='o',linestyle='solid')
# add a title
plt.title("Line Chart Example")
plt.xlabel("Years") # xlabel
plt.ylabel("GDP") # y label
plt.show()
```

```python
from typing import Callable
```

```python
def make_mat(r, c,entry_fn:Callable[[int,int],float]):
    return [[entry_fn(i,j) for j in range(c)]for i in range(r)]
```

```python

def entry_n(i,j):
    return i+j
make_mat(4,5,entry_n)
```
