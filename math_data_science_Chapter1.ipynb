```python
from sympy import *
from sympy.plotting import plot3d
```

```python
x, y = symbols('x,y')
```

```python
f = 2*x + 2*y
f
```

```python
plot3d(f)
```

```python
i, n = symbols('i,n')
summation = Sum(2*i, (i,1,n))
summation
```

```python
summation.subs(n,5).doit() # substitutes n =5
```

```python
def derivative_x(f, x, step_size):
    m = (f(x+step_size)-f(x)) / ((x+step_size)-x)
    return m
```

```python
def my_function(x):
    return x **2  # function like f(x) = x^2
```

```python
slope_at_2 = derivative_x(my_function,2,0.00001)
```

```python
slope_at_2
```

```python
# derivative in sympy
x = Symbol('x')
f = x**3 + 2*x + 19
dx_f = diff(f) # diff() to calculate derivative function 
dx_f
```

```python
dx_f.subs(x,10).doit()
```
