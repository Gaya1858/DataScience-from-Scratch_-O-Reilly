# FIGURE object
### Figure object is used in Matplotlib to represent a graph
#### We can place an instances of Axes inside the figure(graph or canvas). canvas size corresponds to (0,0,1,1) this leaves no space for axis label and ticks. left botton corner (0,0), left upper corner(0,1), right upper corner (1,1) and right lower corner(1,0)
#### So when you create Axes object set size lesser then canvas size . Belw example was created using (0.1,0.1,0.8,0.8)

```python
 # creating some axes points to plot 
x = np.linspace(-5,2,100)
y1 = x**3 +5*x**2 +10 
y2 = 3*x**2 + 10*x 
y3 = 6*x + 10
```

```python
fig, ax = plt.subplots() # creats a below figure and axis object ax
ax.plot(x, y1, color ='blue', label = "y(x)") # each axis object inside the figure
ax.plot(x, y2, color ='green', label = "y'(x)")  # each axis object inside the figure
ax.plot(x, y3, color ='red', label = "y''(x)")  # each axis object inside the figure
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.legend()
```

```python
 fig = plt.figure(figsize = (8,2.5), facecolor= "#f1f1f1",edgecolor = '#a1a1a1') # creating figure object
# figure accpets the figsize with tupleof inch measurement  and the figure canvas facecolor -- 
# all  are optional keywords and if we dont mention it will return with default
#we can add axes to the figure once created
left, bottom, width, height = 0.1,0.1,0.8,0.8
ax = fig.add_axes((left,bottom,width,height), facecolor = "#e1e1e1")# Axes size and facecolor
x = np.linspace(-2,2,1000)
y1 = np.cos(40*x)
y2 = np.exp(-x**2)
ax.plot(x,y1*y2)
ax.plot(x,y2,'g')
ax.plot(x, -y2, 'g')
ax.set_xlabel("X")
ax.set_ylabel("y")
fig.savefig("graph.png", dpi =100, facecolor = "#f1f1f1", edgecolor = '#a1a1a1') # saves the graph as graph.png
```

# Line Properties

```python
x = np.linspace(-5,5,5)
y = np.ones_like(x)
```

```python
def axes_setting(fig, ax, title, ymax):
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_ylim(0,ymax+1)
    ax.set_title(title)
```

```python
fig, ax = plt.subplots(1,4,figsize= (16,3))

# line width
linewidths = [0.5,1.0,2.0,4.0]
for n, linewidth in enumerate(linewidths):
    ax[0].plot(x, y+n, color ='red', linewidth=linewidth)
    axes_setting(fig, ax[0],'linewidth',len(linewidths))
    
#linestyles
linestyles = ['-','-.',':']
for n, linestyle in enumerate(linestyles):
    ax[1].plot(x, y+n, color ='green', lw=2,linestyle=linestyle)
    line, = ax[1].plot(x, y+3, color ='green', lw=2)
    line.set_dashes([10,7,20,7])
    axes_setting(fig, ax[1],'linestyle',len(linestyles)+1)
    
#markers
markers = ['+','o','*','s','.']
for n, marker in enumerate(markers):
    ax[2].plot(x, y+n, color ='orange', lw=1,ls= ':',marker=marker)
    axes_setting(fig, ax[2],'markers',len(markers))
    
```

```python
sym_x = sp.Symbol('x')
x =np.linspace(-2*np.pi,2*np.pi,100)
```

```python
def sin_expansion(x, n):
   # Evaluate nth order Taylor. series expansion of sin(x) for the numerical values in the array x
  
    return sp.lambdify(sym_x,sp.sin(sym_x).series(n =n+1).removeO(),'numpy')(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x,np.sin(x), lw=3,color='red',label='exact')
colors =['blue','green']
linestyles = [':','-.','--']
for idx, n in enumerate(range(1,12,2)):
    ax.plot(x,sin_expansion(x,n),color=colors[idx//3],ls=linestyles[idx%3],lw =1,label= "order %d approx." %(n+1))
    ax.set_ylim(-1.1,1.1)
    ax.set_xlim(-1.5*np.pi,1.5*np.pi)
    
    ax.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad=0.0)
    fig.subplots_adjust(right=.75)
```

# Axis Range

```python
fig, ax = plt.subplots(1,3,figsize=(9,3),subplot_kw={'facecolor':"#ebf5ff"})
# customize your own axis limits
ax[0].plot(x,y,lw=2)
ax[0].set_xlim(-5,35)
ax[0].set_ylim(-1,1)
ax[0].set_title("set_xlim/set_ylim")

# tight axis- for a coordinate range tightly fit the lines it contains 
ax[1].plot(x,y,lw=2)
ax[1].axis('tight')
ax[1].set_title("axis('tight')")

#equal axis - each axis corresponds to the same number of pixels( a ratio preserving coordinate system)
ax[2].plot(x,y,lw=2)
ax[2].axis('equal')
ax[2].set_title("axis('equal')")
```

# Axix Ticks, Tick Labels, and Grids

```python
x = np.linspace(-2*np.pi, 2*np.pi,500)
y = np.sin(x)* np.exp(-x**2/20)
```

```python
fig, ax = plt.subplots(1,4, figsize=(12,3), subplot_kw= {'facecolor':"#ebf5ff"})

ax[0].plot(x,y,lw=2)
ax[0].set_title("default ticks")

ax[1].plot(x,y,lw=2)
ax[1].set_title("set_xticks")
ax[1].set_yticks([-1,0,1])
ax[1].set_xticks([-5,0,5])

ax[2].plot(x,y,lw=2)
ax[2].set_title("set_major_locator")
ax[2].xaxis.set_major_locator(mpl.ticker.MaxNLocator(4))
ax[2].yaxis.set_major_locator(mpl.ticker.FixedLocator([-1,0,1]))
ax[2].xaxis.set_minor_locator(mpl.ticker.MaxNLocator(8))
ax[2].yaxis.set_minor_locator(mpl.ticker.MaxNLocator(8))


ax[3].plot(x,y,lw=2)
ax[3].set_title("set_xticklabels")
ax[3].set_yticks([-1,0,1])
ax[3].set_xticks([-2*np.pi,-np.pi,0,np.pi,2*np.pi])
ax[3].set_xticklabels([r"$-2\pi$",r"$-\pi$",0,r"$\pi$",r"$2\pi$"])

x_minor_ticker= mpl.ticker.FixedLocator([-3 * np.pi/2, -np.pi/2, 0, np.pi/2, 3* np.pi/2])

ax[3].xaxis.set_minor_locator(x_minor_ticker)
ax[3].yaxis.set_minor_locator(mpl.ticker.MaxNLocator(4))
```
